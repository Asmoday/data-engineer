```scala
import org.apache.spark.sql.types.{StructType, StringType}
import org.apache.spark.ml.{Pipeline, PipelineModel}
import org.apache.spark.sql.functions._
import org.apache.spark.sql.DataFrame
```

```scala
println(s"Current spark version is ${spark.version}")
```

```scala
val inputStreamPath = "/home/jovyan/data/events-stream"
val modelPath = "/home/jovyan/models/spark-ml-model"

val dataSchema = new StructType()
    .add("tweet", StringType)

val inputDF = spark
    .readStream
    .schema(dataSchema)
    .option("maxFilesPerTrigger", 1)
    .json(inputStreamPath)
```

```scala
val model = PipelineModel.load(modelPath)
val getProbability = udf((probability: org.apache.spark.ml.linalg.Vector) => probability(1))
```

```scala
val predictionsDF = model.transform(inputDF)

predictionsDF
    .select(getProbability($"probability").alias("clean_probability"))
    .writeStream
    .outputMode("append")
    .format("console")
    .start()
```

```scala
val predictionsDF = model.transform(inputDF)

predictionsDF
    .select(getProbability($"probability").alias("clean_probability"),$"prediction")
    .withColumn("timestamp", current_timestamp)
    .withWatermark("timestamp", "10 seconds")
    .groupBy(window($"timestamp", "10 seconds", "5 second"), $"prediction")
    .count()
    .writeStream
    .outputMode("append")
    .format("console")
    .start()
```

```scala
spark.stop()
```
